```python
import os

# ./my_courses 디렉토리 경로
input_dir = "./my_courses"

# 디렉토리 내 파일 처리
ai_plan_id = 1
ai_plan_detail_id = 1
for file_name in os.listdir(input_dir):
    if file_name.endswith(".txt"):  # .txt 파일만 처리
        # area_code 추출
        area_code = int(file_name.split('_')[0])

        # 파일 읽기 및 eval()로 travel_plans 생성
        file_path = os.path.join(input_dir, file_name)
        with open(file_path, "r", encoding="utf-8") as f:
            file_content = f.read()
            travel_plans = eval(file_content)  # 3중 리스트로 변환

        # 확인용 출력
        # print(f"Processing file: {file_name}")
        # print(f"Area Code: {area_code}")
        # print(f"Travel Plans: {travel_plans}")
    
    for i in range(len(travel_plans)):
        level = i+1
        # ai_plan에 저장
        print(f'id:{ai_plan_id} level : {level}, area_code  : {area_code }')
        ai_plan_id+=1
        for j in range(len(travel_plans[i])):
            attraction_id = travel_plans[i][j][0]
            day = j//5+1
            sequence = j
            # ai_detail_plan에 저장
            print(f'id:{ai_plan_detail_id} day : {day}, sequence  : {sequence }, ai_plan_id : {ai_plan_id}, attraction_id  : {attraction_id }')
            ai_plan_detail_id+=1
```
