```python
with open('./output/attraction_mission.csv','r') as f:
    misssions = f.read().split('\n')
```

```python
from collections import Counter
a = Counter([len(m.split(',')) for m in misssions])
a
```

```python
for i in range(1,len(misssions)-1):
    m = misssions[i]
    misssions[i] = m[:m.find(',')+1]+'\"'+m[m.find(',')+1:-16]+"\""+m[-16:]
```

```python
from collections import Counter
a = Counter([len(m.split(',')) for m in misssions])
a
```

```python
with open('./output/after_attraction_mission.csv','w') as f:
    for i in misssions[:-1]:
        f.write(str(i)+'\n')
```
